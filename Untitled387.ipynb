```python
import os
import json
import yaml
import pickle
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping

import pyswarms as ps

# ============================================================
# PATHS
# ============================================================
BASE_DIR = r"C:\Users\NXTWAVE\Downloads\Student Dropout Risk Prediction"
DATA_PATH = os.path.join(BASE_DIR, "RS_Session_254_AU_352.B.csv")

MODEL_DIR = os.path.join(BASE_DIR, "models")
ARTIFACT_DIR = os.path.join(BASE_DIR, "artifacts")
RESULT_DIR = os.path.join(BASE_DIR, "results")
GRAPH_DIR = os.path.join(BASE_DIR, "graphs")

for d in [MODEL_DIR, ARTIFACT_DIR, RESULT_DIR, GRAPH_DIR]:
    os.makedirs(d, exist_ok=True)

# ============================================================
# LOAD DATA
# ============================================================
df = pd.read_csv(DATA_PATH)
df.columns = df.columns.str.lower()

state_col = "state/ut"
df_states = df[state_col]

df = df.drop(columns=["sl. no.", state_col], errors="ignore")

# ============================================================
# PREVENT DIVISION ERRORS
# ============================================================
df.replace(0, np.nan, inplace=True)

# ============================================================
# DERIVE DROPOUT RISK
# ============================================================
df["drop_p_to_u"] = (
    df["primary level (i - v) - total"] -
    df["upper primary - total"]
) / df["primary level (i - v) - total"]

df["drop_u_to_s"] = (
    df["upper primary - total"] -
    df["secondary (ix-x) - total"]
) / df["upper primary - total"]

df["drop_s_to_h"] = (
    df["secondary (ix-x) - total"] -
    df["higher secondary - total"]
) / df["secondary (ix-x) - total"]

df["dropout_risk"] = df[
    ["drop_p_to_u", "drop_u_to_s", "drop_s_to_h"]
].mean(axis=1)

df.replace([np.inf, -np.inf], np.nan, inplace=True)
df.fillna(df.median(numeric_only=True), inplace=True)
df["dropout_risk"] = df["dropout_risk"].clip(0, 1)

# ============================================================
# FEATURES & TARGET
# ============================================================
X = df.drop(columns=["dropout_risk"])
y = df["dropout_risk"]

# ============================================================
# SCALING
# ============================================================
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

with open(os.path.join(ARTIFACT_DIR, "scaler.pkl"), "wb") as f:
    pickle.dump(scaler, f)

# ============================================================
# LSTM SHAPE
# ============================================================
X_lstm = X_scaled.reshape(X_scaled.shape[0], 1, X_scaled.shape[1])

X_train, X_test, y_train, y_test = train_test_split(
    X_lstm, y, test_size=0.2, random_state=42
)

# ============================================================
# MODEL
# ============================================================
def build_lstm(units, dropout):
    model = Sequential([
        LSTM(int(units), input_shape=(1, X_train.shape[2])),
        Dropout(dropout),
        Dense(1)
    ])
    model.compile(optimizer="adam", loss="mse")
    return model

# ============================================================
# PSO OBJECTIVE
# ============================================================
def pso_objective(particles):
    losses = []
    for p in particles:
        model = build_lstm(int(p[0]), p[1])
        model.fit(X_train, y_train, epochs=8, batch_size=32, verbose=0)
        loss = model.evaluate(X_test, y_test, verbose=0)
        losses.append(loss)
    return np.array(losses)

# ============================================================
# PSO OPTIMIZATION
# ============================================================
optimizer = ps.single.GlobalBestPSO(
    n_particles=8,
    dimensions=2,
    options={"c1": 1.5, "c2": 1.5, "w": 0.7},
    bounds=(np.array([32, 0.1]), np.array([128, 0.5]))
)

best_loss, best_pos = optimizer.optimize(pso_objective, iters=5)

# ============================================================
# FINAL MODEL
# ============================================================
model = build_lstm(int(best_pos[0]), best_pos[1])

model.fit(
    X_train, y_train,
    epochs=40,
    batch_size=32,
    validation_split=0.1,
    callbacks=[EarlyStopping(patience=5, restore_best_weights=True)],
    verbose=1
)

model.save(os.path.join(MODEL_DIR, "dropout_lstm_pso.h5"))

# ============================================================
# PREDICTION
# ============================================================
y_pred = model.predict(X_test).flatten()

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# ============================================================
# SAVE RESULTS
# ============================================================
results = {
    "mse": float(mse),
    "r2_score": float(r2),
    "best_units": int(best_pos[0]),
    "best_dropout": float(best_pos[1])
}

with open(os.path.join(ARTIFACT_DIR, "results.json"), "w") as f:
    json.dump(results, f, indent=4)

with open(os.path.join(ARTIFACT_DIR, "config.yaml"), "w") as f:
    yaml.dump(results, f)

# ============================================================
# CSV + JSON PREDICTIONS
# ============================================================
pred_df = pd.DataFrame({
    "actual_risk": y_test.values,
    "predicted_risk": y_pred
})

pred_df.to_csv(os.path.join(RESULT_DIR, "final_results.csv"), index=False)

with open(os.path.join(RESULT_DIR, "predictions.json"), "w") as f:
    json.dump(pred_df.to_dict(orient="records"), f, indent=4)

# ============================================================
# ===================== GRAPHS ===============================
# ============================================================

# 1️⃣ Accuracy / R2 Graph
plt.figure()
plt.bar(["R2 Score"], [r2])
plt.title("Model Performance (R²)")
plt.savefig(os.path.join(GRAPH_DIR, "accuracy_r2.png"))
plt.show()

# 2️⃣ Comparison Graph
plt.figure()
plt.plot(y_test.values, label="Actual")
plt.plot(y_pred, label="Predicted")
plt.legend()
plt.title("Actual vs Predicted Dropout Risk")
plt.savefig(os.path.join(GRAPH_DIR, "comparison_graph.png"))
plt.show()

# 3️⃣ Risk Distribution Graph
plt.figure()
sns.histplot(df["dropout_risk"], bins=10, kde=True)
plt.title("Dropout Risk Distribution")
plt.savefig(os.path.join(GRAPH_DIR, "risk_distribution.png"))
plt.show()

# 4️⃣ Heatmap (State-wise)
heat_df = pd.DataFrame({
    "State": df_states,
    "Risk": df["dropout_risk"]
})

plt.figure(figsize=(8, 10))
heat_df = heat_df.sort_values("Risk")
sns.heatmap(
    heat_df[["Risk"]],
    yticklabels=heat_df["State"],
    cmap="Reds",
    cbar=True
)
plt.title("State-wise Dropout Risk Heatmap")
plt.savefig(os.path.join(GRAPH_DIR, "risk_heatmap.png"))
plt.show()

print("\n✅ FULL PIPELINE EXECUTED SUCCESSFULLY")
print("📁 All results, graphs, CSV, JSON saved")
```
