```python
import os
import json
import yaml
import pickle
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping

import pyswarms as ps

# ============================================================
# PATHS
# ============================================================
BASE_DIR = r"C:\Users\NXTWAVE\Downloads\Student Dropout Risk Prediction"
DATA_PATH = os.path.join(BASE_DIR, "RS_Session_254_AU_352.B.csv")

MODEL_DIR = os.path.join(BASE_DIR, "models")
ARTIFACT_DIR = os.path.join(BASE_DIR, "artifacts")
RESULT_DIR = os.path.join(BASE_DIR, "results")
GRAPH_DIR = os.path.join(BASE_DIR, "graphs")

for d in [MODEL_DIR, ARTIFACT_DIR, RESULT_DIR, GRAPH_DIR]:
    os.makedirs(d, exist_ok=True)

# ============================================================
# LOAD & CLEAN DATA
# ============================================================
df = pd.read_csv(DATA_PATH)
df.columns = df.columns.str.lower()

states = df["state/ut"]
df = df.drop(columns=["sl. no.", "state/ut"], errors="ignore")

df.replace(0, np.nan, inplace=True)

# ============================================================
# DERIVE DROPOUT RISK (TARGET)
# ============================================================
df["d1"] = (df["primary level (i - v) - total"] - df["upper primary - total"]) / df["primary level (i - v) - total"]
df["d2"] = (df["upper primary - total"] - df["secondary (ix-x) - total"]) / df["upper primary - total"]
df["d3"] = (df["secondary (ix-x) - total"] - df["higher secondary - total"]) / df["secondary (ix-x) - total"]

df["dropout_risk"] = df[["d1", "d2", "d3"]].mean(axis=1)

df.replace([np.inf, -np.inf], np.nan, inplace=True)
df.fillna(df.median(numeric_only=True), inplace=True)
df["dropout_risk"] = df["dropout_risk"].clip(0, 1)

# ============================================================
# FEATURES & TARGET
# ============================================================
X = df.drop(columns=["dropout_risk"])
y = df["dropout_risk"]

scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

with open(os.path.join(ARTIFACT_DIR, "pis_scaler.pkl"), "wb") as f:
    pickle.dump(scaler, f)

X_lstm = X_scaled.reshape(X_scaled.shape[0], 1, X_scaled.shape[1])

X_train, X_test, y_train, y_test = train_test_split(
    X_lstm, y, test_size=0.2, random_state=42
)

# ============================================================
# LSTM MODEL
# ============================================================
def build_lstm(units, dropout):
    model = Sequential([
        LSTM(int(units), input_shape=(1, X_train.shape[2])),
        Dropout(dropout),
        Dense(1)
    ])
    model.compile(optimizer="adam", loss="mse")
    return model

# ============================================================
# FITNESS FUNCTION
# ============================================================
def fitness(params):
    units, dropout = int(params[0]), params[1]
    model = build_lstm(units, dropout)
    model.fit(X_train, y_train, epochs=8, batch_size=32, verbose=0)
    return model.evaluate(X_test, y_test, verbose=0)

# ============================================================
# AIS INITIAL POPULATION
# ============================================================
ais_population = np.array([
    [random.randint(32, 128), random.uniform(0.1, 0.5)]
    for _ in range(8)
])

# ============================================================
# PSO OPTIMIZER (USED AFTER AIS SELECTION)
# ============================================================
bounds = (
    np.array([32, 0.1]),
    np.array([128, 0.5])
)

optimizer = ps.single.GlobalBestPSO(
    n_particles=ais_population.shape[0],
    dimensions=2,
    options={"c1": 1.5, "c2": 1.5, "w": 0.7},
    bounds=bounds,
    init_pos=ais_population
)

best_loss, best_pos = optimizer.optimize(
    lambda p: np.array([fitness(x) for x in p]),
    iters=5
)

best_units = int(best_pos[0])
best_dropout = float(best_pos[1])

# ============================================================
# FINAL MODEL TRAINING
# ============================================================
final_model = build_lstm(best_units, best_dropout)

final_model.fit(
    X_train, y_train,
    epochs=40,
    batch_size=32,
    validation_split=0.1,
    callbacks=[EarlyStopping(patience=5, restore_best_weights=True)],
    verbose=1
)

final_model.save(os.path.join(MODEL_DIR, "pis_dropout_lstm_ais_pso.h5"))

# ============================================================
# EVALUATION
# ============================================================
y_pred = final_model.predict(X_test).flatten()

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# ============================================================
# SAVE RESULTS
# ============================================================
results = {
    "mse": float(mse),
    "r2_score": float(r2),
    "best_units": best_units,
    "best_dropout": best_dropout
}

with open(os.path.join(ARTIFACT_DIR, "pis_results.json"), "w") as f:
    json.dump(results, f, indent=4)

with open(os.path.join(ARTIFACT_DIR, "pis_config.yaml"), "w") as f:
    yaml.dump(results, f)

pred_df = pd.DataFrame({
    "actual_risk": y_test.values,
    "predicted_risk": y_pred
})

pred_df.to_csv(os.path.join(RESULT_DIR, "pis_final_results.csv"), index=False)

with open(os.path.join(RESULT_DIR, "pis_predictions.json"), "w") as f:
    json.dump(pred_df.to_dict(orient="records"), f, indent=4)

# ============================================================
# =================== GRAPHS ===============================
# ============================================================

plt.figure()
plt.bar(["R2 Score"], [r2])
plt.title("PIS Hybrid Model Performance (AIS + PSO)")
plt.savefig(os.path.join(GRAPH_DIR, "pis_accuracy_r2.png"))
plt.show()

plt.figure()
plt.plot(y_test.values, label="Actual")
plt.plot(y_pred, label="Predicted")
plt.legend()
plt.title("PIS Actual vs Predicted")
plt.savefig(os.path.join(GRAPH_DIR, "pis_comparison_graph.png"))
plt.show()

plt.figure()
sns.histplot(df["dropout_risk"], bins=10, kde=True)
plt.title("PIS Dropout Risk Distribution")
plt.savefig(os.path.join(GRAPH_DIR, "pis_risk_distribution.png"))
plt.show()

heat_df = pd.DataFrame({"State": states, "Risk": df["dropout_risk"]}).sort_values("Risk")

plt.figure(figsize=(8, 10))
sns.heatmap(
    heat_df[["Risk"]],
    yticklabels=heat_df["State"],
    cmap="Reds",
    cbar=True
)
plt.title("PIS State-wise Dropout Risk Heatmap")
plt.savefig(os.path.join(GRAPH_DIR, "pis_risk_heatmap.png"))
plt.show()

print("\n✅ HYBRID AIS + PSO (PIS) PIPELINE COMPLETED SUCCESSFULLY")
print("📁 All outputs saved with 'pis_' prefix")
```
